```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y = make_regression(n_samples = 100,n_features = 1, n_targets = 1, n_informative = 1,noise = 60, random_state = 42)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(X,y)
print(model1.coef_, model1.intercept_)
```

```python
from sklearn.linear_model import Ridge
model2 = Ridge(alpha = 10)
model2.fit(X,y)
print(model2.coef_, model2.intercept_)
```

```python
from sklearn.linear_model import Ridge
model3 = Ridge(alpha = 100)
model3.fit(X,y)
print(model3.coef_, model3.intercept_)
```

```python
plt.scatter(X,y)
plt.plot(X, model1.predict(X), color = 'r', label = 'LinearRegression')
plt.plot(X, model2.predict(X), color = 'b', label = 'alpha = 10')
plt.plot(X, model3.predict(X), color = 'g', label = 'alpha = 100')
plt.legend()
```

```python
class MyRidge:

  def __init__(self, alpha = 1):
    self.alpha = alpha
    self.m = None
    self.b = None
  
  def fit(self, X_train, y_train):

    num = 0
    den = 0

    for i in range(X_train.shape[0]):

      num = num + ((y_train[i] - y_train.mean()) * (X_train[i]-X_train.mean()))
      den = den + ((X_train[i]-X_train.mean())**2)

    self.m = num/(den +self.alpha)
    self.b = y_train.mean() - (self.m*X_train.mean())
    print(self.m, self.b)


  def predict(self, X_test):
    return self.m*X_test + self.b
```

```python
ridge = MyRidge(alpha = 100)
```

```python
ridge.fit(X, y)
```
