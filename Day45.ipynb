```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Manish-k723/100-days-of-ML/main/train.csv', usecols = ['Age','Pclass', 'SibSp','Parch','Survived'])
```

```python
df.head()
```

```python
df.dropna(inplace = True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y, scoring = 'accuracy', cv = 6))
```

Applying Feature- construction

```python
X['family_size'] = X['Parch']+X['SibSp']+1
```

```python
X.head()
```

```python
def family(num):
  if num==1:
    return 0
  elif num>1 and num<=4:
    return 1
  else:
    return 2
```

```python
X['family_type'] = X['family_size'].apply(family)
```

```python
X.drop(['SibSp',	'Parch'	,'family_size'], axis = 1)
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring = 'accuracy',cv = 10))
```
