```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples = 4, noise = 100, n_features = 1, n_informative = 1, random_state = 15, n_targets = 1)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
#OLS
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
model.coef_, model.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X, model.predict(X), color = 'red')
```

```python
# Let's apply gradient descent assuming slope is contant and its value is 170.58
# and initial random values of intercept is, b =0 
y_pred = ((170.58*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X, model.predict(X), color = 'red')
plt.plot(X, y_pred, color = 'blue')
plt.legend()
```

```python
m = 170.58
b = 0

loss_slope = -2*np.sum(y- (m*X.ravel())-b)
loss_slope
```

```python
lr = 0.1

step_size = lr*loss_slope
step_size
```

```python
b = b- step_size
b
```

```python
y_pred1 = ((170.58*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X, model.predict(X), color = 'red', label = 'OLS')
plt.plot(X, y_pred, color = 'blue', label = f'b = {0}')
plt.plot(X, y_pred1, color = 'black', label = f'b = {b}')
plt.legend()
```

```python
# Iteration:2
loss_slope = -2*np.sum(y- (m*X.ravel())-b)
step_size = lr*loss_slope
b = b-step_size
b
```

```python
y_pred2 = ((170.58*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X, model.predict(X), color = 'red', label = 'OLS')
plt.plot(X, y_pred, color = 'blue', label = f'b = {0}')
plt.plot(X, y_pred1, color = 'purple', label = f'b = {-34.35}')
plt.plot(X, y_pred2, color = 'black', label = f'b = {b}')
plt.legend()
plt.show()
```

```python
# Iteration:3
loss_slope = -2*np.sum(y- (m*X.ravel())-b)
step_size = lr*loss_slope
b = b-step_size
b
```

```python
y_pred3 = ((170.58*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X, model.predict(X), color = 'red', label = 'OLS')
plt.plot(X, y_pred, color = 'blue', label = f'b = {0}')
plt.plot(X, y_pred1, color = 'purple', label = f'b = {-34.35}')
plt.plot(X, y_pred2, color = 'brown', label = f'b = {-41.42}')
plt.plot(X, y_pred3, color = 'black', label = f'b = {b}')
plt.legend()
plt.show()
```

```python
# Let's make a iterative function for it
m = 170.58
b = 300
lr = 0.01

epochs = 100

for i in range(epochs):
  loss_slope = -2*np.sum(y- (m*X.ravel())-b)
  step_size = lr*loss_slope
  b = b-step_size
  y_pred = ((170.58*X)+b).reshape(4)

  plt.plot(X,y_pred)

plt.scatter(X,y, color = 'gold')
```
